```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
a=cv2.imread("/content/taj mahal.jpg")
```

```python
cv2_imshow(a)
```

```python
# converting original image(BGR scale) to gray scale
gray=cv2.cvtColor(a,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray)
```

```python
#inverting the grey image
invert=cv2.bitwise_not(gray)
```

```python
cv2_imshow(invert)
```

```python
#getting the blurred inverted image
blur=cv2.GaussianBlur(invert,(21,21),0)
```

```python
cv2_imshow(blur)
```

```python
def sketch(i,j):
  return cv2.divide(i,255-j,scale=256)
```

```python
final=sketch(gray,blur)
```

```python
cv2_imshow(final)
```
